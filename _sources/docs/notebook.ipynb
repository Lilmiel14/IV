## Dataset and preprocessing

### 1 Average screentime 
The first dataset will show the average screen time of 45 countries. The dataset includes information such as time spent on devices, time spent on social media and the usage of the internet via mobiles, computers, etc. Worldwide, the average person has a screentime of 6 hours and 37 minutes. This dataset provides an opportunity to analyze the digital habits and trends across different countries. The initial dataset was obtained from kaggle.com in the form of a CSV file. In order to appropriately preprocess the data, the time variable was converted into a decimal value. Both the variables 'Time Spent on Devices' and 'Time Spent using Social Media' are graphically represented in a bar chart.
 

### 2 Effects of social media
The second dataset contains data about the usage of social media by students. Information about the students’ habits and preferences are provided, such as their preferred social media platform, the amount of time they spend on social media and their preferred communication methods. The dataset provides opportunities to analyze the impact of social media on students’ lives. The second dataset, sourced from data.world, was provided as an XLSX file. Prior to analysis, the data was already preprocessed to illustrate the occurrences of various types of cybercrime victims and the favored mode of communication. Both aspects are presented individually through two distinct donut charts. Since participants were allowed to select multiple options, the occurrences of each option were aggregated. Using the same dataset employed for the donut charts, we further processed the data to showcase the daily levels of physical activity and social interaction. These statistics are visually depicted using two pie charts.

### 3 Factors contributing to internet addiction. 
The third dataset provides data about factors contributing to internet addiction across a diverse age group. The main contributing factors were found to be a high use of discussion forums, a lack of self care and engaging in online video gaming. By analyzing the different factors that influence internet addiction across different age groups, it’s possible to understand the underlying problem. The third dataset was discovered on figshare.com in the form of a SAV file. For appropriate preprocessing, the columns in the file were renamed to enhance the overall presentation, for instance, changing 'DASS21_depression' to 'depression'. Subsequently, a box plot was generated to visualize the anxiety, depression, and stress levels of individuals who engage in social media usage.



### 4 Fake news
Social media has become the biggest platform for sharing different types of content, including news articles. The fourth dataset provides information about the authenticity of articles from different websites. The dataset contains variables such as author, publish date and label (real or fake). The dataset was found on kaggle.com in the form of a CSV file. Before creating the plot, the rows where the language was missing were removed. Furthermore, the values in the column "hasImage" were changed from binary to "Yes" and "No" for easier readability. With this data, a parallel categories plot was created, with the line color based on the "Real" and "Fake" labels.

### 5 Social media privacy awareness
The fifth dataset provides data about a survey conducted on 127 students at Institut Teknologi Sepuluh Nopember (ITS) Surabaya about their social media awareness. Variables such as age and sex are present in the dataset, as well as questions about the students’ social media usage, their personal information and privacy permissions. The dataset was found on zenodo.org in the form of an XLSX file. The language of the dataset was Indonesian, so the language was converted to English. Two new data frames were created from this dataset. The first one being questions that could be answered with "Yes" and "No" and the second being the degree of concern regarding social media. Both of the data frames were plotted using the seaborn library.



## The Bright Side of Social Media
For the bright side of social media we have two arguments which will be analyzed and supported by subarguments and two visualizations. The first visualization supports the argument that social media increases the awareness of students. The second visualization supports the argument that social media revolutionizes communication.


```python
import pandas as pd
import plotly.graph_objs as go
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
import textwrap
import openpyxl
```

```python
{
    "tags": [
        "hide-input",
    ]
}

df = pd.read_csv('screen.csv')

def time_to_decimal(time_str):
    hours, minutes = map(int, time_str.split(':'))
    decimal_hours = hours + minutes / 60
    return decimal_hours

df['Total Time Spent on Devices'] = df['Total Time Spent on Devices'].apply(time_to_decimal)
df['Time Spent using Social Media'] = df['Time Spent using Social Media'].apply(time_to_decimal)

countries = ['Netherlands', 'Worldwide', 'Brazil', 'Russia', 'USA', 'Indonesia', 'China', 'Australia', 'South Africa', 'UK', 'France', 'Germany', 'Saudi Arabia', 'India', 'Japan']

df = df[df.Country.isin(countries)]

trace1 = go.Bar(
    x=df['Country'],
    y=df['Total Time Spent on Devices'],
    name='Total Time Spent on Devices',
    marker=dict(color='#FFAC1C')
)
trace2 = go.Bar(
    x=df['Country'],
    y=df['Time Spent using Social Media'],
    name='Time Spent using Social Media',
    marker=dict(color='DodgerBlue'))


data = [trace1, trace2]

layout = go.Layout(
    title='Total time spent on screentime and social media worldwide',
    xaxis=go.layout.XAxis(
        type='category' ,
        title='Country'
    ),
    yaxis=go.layout.YAxis(
        tickformat=',.Of',
        title='Time (hours)' 
    ),
    height=400,
    barmode='overlay'
)

fig = go.Figure(data=data, layout=layout)
fig.show()

print('The Pearson correlation between the two variables is:', df['Total Time Spent on Devices'].corr(df['Time Spent using Social Media']))
```

*Figure 1: This chart shows the total time spent on screentime and social media worldwide. The screentime of each country is different and the time spent using social media is a significant part of the total time spent on devices. Moreover, the Pearson correlation between the total spent on screentime and social media is around 0,94. This indicates that there is a strong correlation between these variables.*

Social media increases the awareness of students. First of all, social media is accessible for plenty of students. Social media applications are easily accessible through various devices. This accessibility ensures that students can stay connected and informed regardless of their location or time zone. This accessibility is also proven by our first chart. The chart shows that the time students spend using social media is a significant part of the total time spent on devices.

## The Dark Side of Social Media
For the dark side of social media we have three arguments which will be analyzed and supported by four visualizations. The first two visualizations support the argument that social media can lead to addiction and health problems. The third visualization supports the argument that students can be more exposed to fake news. The last visualization supports the argument that students are concerned with their privacy on social media. 

```python
{
    "tags": [
        "hide-input",
    ]
}

df = pd.read_spss('addiction.sav')

df.rename(columns={'DASS21_depression': 'Depression', 'DASS21_anxiety': 'Anxiety', 'DASS21_stress': 'Stress'}, inplace=True)

order = ['Never', 'Seldom', 'Sometimes', 'Often', 'Very Often']

fig3 = px.box(df, y=['Anxiety', 'Depression', 'Stress'],
             labels={'variable': '', 'value': 'Scale'}, 
             title='Relation between DASS Scale and Social Network Usage', facet_col='Social_Network', 
             color=df['Social_Network'], category_orders={'Social_Network': order})
fig3.for_each_annotation(lambda a: a.update(text=a.text.split("=")[1]))

fig3.show()
```

*Figure 3: The figure presented above provides a visual about the relationship between the DASS (Depression, Anxiety and Stress Scale) and social network usage. The DASS scale is a psychological analysis tool to measure the severity of these three mental health conditions. Social network usage refers to the frequency to which the participants use online platforms to communicate. The network usage is ordered in an ordinal scale, ranging from the lowest level (‘Never’) to the highest level (‘Very often’). Information about the DASS is displayed in five different boxplots: one boxplot for each level.*

A big problem with social media is that on the platforms the users will compare their lives to others, which can lead to low self-esteem, depression and anxiety. By creating such an environment, where individuals are constantly exposed to other people’s lives, the social media users can develop feelings of dissatisfaction with their own lives. As seen in the ‘Very often’ boxplot, all mental health conditions have a relatively max score. So the people who spend a lot of time using social media experience more mental health problems.

```python
{
    "tags": [
        "hide-input",
    ]
}

df5 = pd.read_csv('fakenews.csv')
df5.head()

df5 = df5[df5['language'].notna()]
df5['hasImage'] = df5['hasImage'].apply(lambda x: 'Yes' if x == 1 else 'No')

language = ['english', 'german', 'french', 'spanish', 'ignore']
type = ['bs', 'bias', 'conspiracy', 'hate', 'satire', 'state', 'junksci', 'fake']

fig_parallel2 = go.Figure(data=go.Parcats(
    dimensions=[
        {'label': 'Language', 'values': df5['language'], 'categoryarray': language},
        {'label': 'Type', 'values': df5['type'], 'categoryarray': type},
        {'label': 'Has Image', 'values': df5['hasImage']},
        {'label': 'Real or Fake', 'values': df5['label']}
    ],
    line={'color': df5['label'].apply(lambda x: 'mediumseagreen' if x == 'Real' else '#E74739')}
))

fig_parallel2.update_layout(title='Analysis of Fake News Classification')
fig_parallel2.show()
```

*Figure 5: This visualization consists of a parallel categories diagram about the classification of fake news. The diagram contains four different variables: ‘Language’, ‘Type’, ‘Has image’, ‘Real or Fake’. The ‘Language’ variable represents the language of the articles, the ‘Type’ variable represents the type of the articles, the ‘Has image’ variable represents whether the article has an image and the ‘Real or Fake’ variable represents the classification of the articles as either ‘real’ or ‘fake’.*

```python
{
    "tags": [
        "hide-input",
    ]
}

df3 = pd.read_excel('awareness.xlsx')

# Translate questions
df3.rename(columns={'Umur': 'Age', 'Jenis Kelamin': 'Sex', 
                   'Seberapa sering menggunakan media sosial? (jam/hari)': 'How often do you use social media? (hours/day)', 
                   'Berapa tahun telah menggunakan media sosial?': 'How many years have you used social media?', 
                   'Apakah informasi pribadi yang ada di media sosial Anda adalah informasi yang sebenarnya? ': 'Is personal information on your social media real information?',  
                   'Apakah informasi data diri dan kegiatan Anda yang ada di media sosial Anda terbuka untuk umum? ': 'Is the information about yourself and your activities on your social media open to the public?', 
                   'Apakah Anda memanfaatkan pengaturan (setting) privasi di media sosial? ': 'Do you use privacy settings on social media?',   
                   'Apakah Anda hanya menambahkan orang lain sebagai teman di media sosial, hanya jika Anda mengenal mereka? ': 'Do you only add other people as friends on social media, only if you know them?',  
                   'Apakah Anda membaca prosedur privacy and policy sebelum sign up ke media sosial? ': 'Did you read the privacy and policy procedures before signing up for social media?',   
                   'Apakah Anda merasa khawatir (concern) terhadap privacy and policy media sosial? ': 'Do you feel worried (concern) about the privacy and policy of social media?', 
                   'Apakah Anda mengetahui izin akses (permission access) media sosial terhadap perangkat (smartphone, tablet, dll.) Anda?': 'Do you know the social media access permissions for your device (smartphone, tablet, etc.)?',
                   'Apakah Anda merasa khawatir (concern) terhadap izin akses media sosial terhadap perangkat Anda? ': 'Are you concerned about social media access permissions on your device?', 
                   'Apakah Anda selalu mengizinkan akses media sosial terhadap perangkat Anda? ': 'Do you always allow social media access on your device?', 
                   'Apakah Anda merasa khawatir (concern) terhadap akses pihak ketiga (third-party access) di media sosial? ': 'Are you concerned about third-party access on social media?', 
                   'Apakah Anda selalu mengizinkan akses pihak ketiga di media sosial? ': 'Do you always allow third party access on social media?', 
                   'Manakah hal di bawah ini yang pernah Anda bagikan di media sosial? ': 'Which of the following have you shared on social media?', 
                   'Apakah Anda khawatir (concern) terhadap privasi di media sosial? ': 'Are you concerned about privacy on social media?',
                   'Apa yang Anda pahami tentang “privasi” dalam media sosial? ': 'What do you understand about “privacy” in social media?'}, inplace=True)

# Translate Sex, Yes and No
df3['Sex'] = df3['Sex'].apply(lambda x: 'Male' if x == 'Laki-laki' else 'Female')
df3['Is personal information on your social media real information?'] = df3['Is personal information on your social media real information?'].apply(lambda x: 'Yes' if x == 'Ya' else 'No') 
df3['Is the information about yourself and your activities on your social media open to the public?'] = df3['Is the information about yourself and your activities on your social media open to the public?'].apply(lambda x: 'Yes' if x == 'Ya' else 'No') 
df3['Do you use privacy settings on social media?'] = df3['Do you use privacy settings on social media?'].apply(lambda x: 'Yes' if x == 'Ya' else 'No')
df3['Do you only add other people as friends on social media, only if you know them?'] = df3['Do you only add other people as friends on social media, only if you know them?'].apply(lambda x: 'Yes' if x == 'Ya' else 'No') 
df3['Did you read the privacy and policy procedures before signing up for social media?'] = df3['Did you read the privacy and policy procedures before signing up for social media?'].apply(lambda x: 'Yes' if x == 'Ya' else 'No') 
df3['Do you know the social media access permissions for your device (smartphone, tablet, etc.)?'] = df3['Do you know the social media access permissions for your device (smartphone, tablet, etc.)?'].apply(lambda x: 'Yes' if x == 'Ya' else 'No') 

# Translate how worried people are
value_mapping = {'Sangat khawatir': 'Very worried', 'Cukup khawatir': 'Moderately worried', 'Tidak terlalu khawatir': 'Slightly worried', 'Tidak khawatir sama sekali': 'Not worried'}
df3['Do you feel worried (concern) about the privacy and policy of social media?'] = df3['Do you feel worried (concern) about the privacy and policy of social media?'].replace(value_mapping)
df3['Are you concerned about social media access permissions on your device?'] = df3['Are you concerned about social media access permissions on your device?'].replace(value_mapping)
df3['Are you concerned about third-party access on social media?'] = df3['Are you concerned about third-party access on social media?'].replace(value_mapping)
df3['Are you concerned about privacy on social media?'] = df3['Are you concerned about privacy on social media?'].replace(value_mapping)

# Translate Always and not always to Yes and No
value_mapping2 = {'Tidak Selalu': 'No', 'Selalu': 'Yes'}
df3['Do you always allow social media access on your device?'] = df3['Do you always allow social media access on your device?'].replace(value_mapping2)
df3['Do you always allow third party access on social media?'] = df3['Do you always allow third party access on social media?'].replace(value_mapping2)

# Make plots
df4 = pd.melt(df3, value_vars=['Is personal information on your social media real information?', 'Is the information about yourself and your activities on your social media open to the public?',
                               'Do you know the social media access permissions for your device (smartphone, tablet, etc.)?', 'Do you always allow social media access on your device?', 
                               'Do you always allow third party access on social media?', 'Did you read the privacy and policy procedures before signing up for social media?',
                               'Do you use privacy settings on social media?', 'Do you only add other people as friends on social media, only if you know them?'],
                                var_name='Survey questions', value_name='Answer')
wrapped_labels = [textwrap.fill(label, 24) for label in df4['Survey questions']]
plt.figure(figsize=(25, 8))
sns.set(font_scale=1.25)
sns.countplot(x=wrapped_labels, hue='Answer', data=df4, width=0.7, palette = ['#1565C0', '#D32F3F']).set(title = 'Social media survey results')
plt.show()

df5 = pd.melt(df3, value_vars=['Do you feel worried (concern) about the privacy and policy of social media?', 'Are you concerned about social media access permissions on your device?',
                               'Are you concerned about third-party access on social media?', 'Are you concerned about privacy on social media?'],
                                var_name='Survey questions', value_name='Degree')
wrapped_labels2 = [textwrap.fill(label, 24) for label in df5['Survey questions']]
plt.figure(figsize=(25, 8))
sns.countplot(x=wrapped_labels2, hue='Degree', data=df5, width=0.7, palette = ["#FF5C5C", "#FF8A42", "#FFC26B", "#83FF73"]).set(title = 'Degree of concern regarding social media')
plt.show()
```

The second questionnaire shows that students are concerned about their privacy on social media. For example, students are worried about the social media access permissions on their phone.*
